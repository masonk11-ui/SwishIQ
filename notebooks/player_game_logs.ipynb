```python
import pandas as pd
import numpy as np
```

```python
from nba_api.stats.endpoints import playergamelogs

def get_player_game_logs():
    df = playergamelogs.PlayerGameLogs(
        season_nullable="2025-26",
        season_type_nullable="Playoffs",
        timeout=60
    ).get_data_frames()[0]


    return df
```

```python
df = get_player_game_logs()
```
