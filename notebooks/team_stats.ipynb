```python
import pandas as pd
import numpy as np
from nba_api.stats.endpoints import teamgamelogs
pd.set_option('display.max_columns', None)
```

```python
team_game_logs = teamgamelogs.TeamGameLogs(
    season_nullable='2025-26',
    season_type_nullable= 'Playoffs',
    timeout= 60
)
```

```python
playoff_game_logs = team_game_logs.get_data_frames()[0]
cavs_game_logs = playoff_game_logs[playoff_game_logs['TEAM_ABBREVIATION'] == 'CLE']
cavs_game_logs
```

```python
playoff_game_logs['PTSA'] = playoff_game_logs['PTS'] - playoff_game_logs['PLUS_MINUS']
stats = ['PTS', 'PTSA', 'FG_PCT', 'FG3_PCT', 'FT_PCT', 'OREB',
         'DREB', 'REB', 'AST', 'TOV', 'STL', 'BLK',
         'PLUS_MINUS']


team_stats_df = (
    playoff_game_logs
    .groupby("TEAM_NAME")[stats]
    .mean().round(2)
    .reset_index()
)
```

```python
team_stats_df
```

```python
team_stats_df = team_stats_df.groupby('TEAM_NAME').agg({
    'PTS' : 'sum',
    'PTSA' : 'sum',
    'Possessions' : 'sum'
}).reset_index().rename(columns={
        "TEAM_NAME": "Team"})

team_stats_df['O_RTG'] = 100 * team_stats_df['PTS'] / team_stats_df['Possessions']
team_stats_df['D_RTG'] = 100 * team_stats_df['PTSA'] / team_stats_df['Possessions']
team_stats_df['NET_RTG'] = team_stats_df['O_RTG'] - team_stats_df['D_RTG']
```
